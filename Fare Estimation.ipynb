```python
# Distance generator
from math import cos, sin, sqrt , ceil , atan2 , radians
def distance(lat1, lon1, lat2, lon2):
    # approximate radius of earth in km
    R = 6373.0
    lat1 = radians(lat1)
    lon1 = radians(lon1)
    lat2 = radians(lat2)
    lon2 = radians(lon2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return R * c

distances = np.array([ distance( latitudes1[_] , longitudes1[_] , latitudes2[_] , longitudes2[_] ) for _ in range(0,samples) ])
dist = pd.Series(distances)
distance_max = int( ceil( dist.max() ) ) + 10
distances = [ np.random.randint( dist[i] + 2 , distance_max ) for i in range(0,samples) ]
pd.DataFrame( { 'min_distance' : dist ,
                'distances' : distances } ).boxplot()
dist_describe = pd.DataFrame(distances).describe().T
print dist_describe['75%']
```

```python
# Fare generator
# Cost generation
# 0 - 25 % to min fare class( using 1st quartile ) 
# 75 % - 100 % to max fare class ( using 3rd quartile )
# Morning 9:00 to 20:59  , Night 21:00 PM to 8:59
import time
fares = []
for i in range(0,samples) :    #    mode        min        average    max        base_fare            wait_time_fare        traffic_wait_time_fare    luggage_fare   day_night fare 
    fare = 0.0
    if mode_sns[i] == 0 :      #    Metro       0.8333     0.9267     1          5                    0                     0                         0              0
        fare = 5.0   # base fare
        if distances[i] <= int( round(dist_describe['25%'],0) ) :
            fare = fare + distances[i]*1
        elif distances[i] < int( round(dist_describe['75%'],0) ) :
            fare = fare + distances[i]*0.9267
        else :
            fare = fare + distances[i]*0.8333
    elif mode_sns[i] == 1 :        #    Bus         0.5263     0.9444     1.6667     8                    0                     0                         0              0
        fare = 8.0   # base fare
        if distances[i] <= int( round(dist_describe['25%']) ) :
            fare = fare + distances[i]*1.6667
        elif distances[i] < int( round(dist_describe['75%']) ) :
            fare = fare + distances[i]*0.9444
        else :
            fare = fare + distances[i]*0.5263
    elif mode_sns[i] == 2 :        #    Minibus     0.5625     1.1323     2          15                   0                     0                         0              0                 
        fare = 15.0
        if distances[i] <= int( round(dist_describe['25%']) ) :
            fare = fare + distances[i]*2
        elif distances[i] < int( round(dist_describe['75%']) ) :
            fare = fare + distances[i]*1.1323
        else:
            fare = fare + distances[i]*0.5625
    elif mode_sns[i] == 3 :        #    AC bus      0.6316     1.1333     2          18                    0                    0                          0             0
        fare = 18.0
        if distances[i] <= int( round(dist_describe['25%']) ) :
            fare = fare + distances[i]*2
        elif distances[i] < int( round(dist_describe['75%']) ) :
            fare = fare + distances[i]*1.1333
        else :
            fare = fare + distances[i]*0.6316

#    Minimum taxi fare for 0-2 km distance(journey): Rs 25/-
#    For every 1km distance the taxi fare is increase by Rs 12.00/-
#    For every 1 minute after initial 5 minutes, the taxi waiting charge is Rs 0.5/- per min
#    There will be Rs 4/- per kg extra for carrying luggage that weighs beyond 20kgs
#    AC Taxi fare is +20% of non-AC fare
#    Fare is increased by +10% at night
    elif mode_sns[i] == 4 :       #     Non - AC Taxi
        fare = 25.0
        if distances[i] > 2:
            fare = fare + ( distances[i] - 2 ) * 12
        if wait_times[i] + traffic_wait_times[i] > 5 :
            fare = fare + ( wait_times[i] + traffic_wait_times[i] - 5 ) * 0.5
        timestamp = time.strptime( timestamps[i] , "%Y-%m-%d %H:%M:%S" )
        if timestamp.tm_hour >= 21 or timestamp.tm_hour < 9 :
            fare = fare*1.1
    elif mode_sns[i] == 5 :      #     AC Taxi
        fare = 25.0
        if distances[i] > 2:
            fare = fare + ( distances[i] - 2 ) * 12
        if wait_times[i] + traffic_wait_times[i] > 5 :
            fare = fare + ( wait_times[i] + traffic_wait_times[i] - 5 ) * 0.5
        timestamp = time.strptime( timestamps[i] , "%Y-%m-%d %H:%M:%S" )
        if timestamp.tm_hour >= 21 or timestamp.tm_hour < 9 :
            fare = fare*1.1
        fare = fare*1.2
    elif mode_sns[i] == 6 :     #      Auto rikshaw      min. - 8   avg. - 9  max. - 11  Night fare increased by 0.06
        if distances[i] <= int( round(dist_describe['25%']) ):
            fare = distances[i]*8
        elif distances[i] < int( round(dist_describe['75%']) ) :
            fare = distances[i]*9
        else :
            fare = distances[i]*11
        timestamp = time.strptime( timestamps[i] , "%Y-%m-%d %H:%M:%S" )
        if timestamp.tm_hour >= 21 or timestamp.tm_hour < 9 :
            fare = fare*1.4286
    fares.append( fare )
#Fare levels
# SN    City        Fare Index
# 0     Mumbai      1.1776
# 1     Delhi       1.1662
# 2     Bangalore   1.0835
# 3     Chennai     1.0275
# 4     Kolkata     1.0000
# 5     Varanasi    1.0089
for i in range(0,samples):
    if cities[i] == 0 :
        fares[i] = round( 1.333333 * fares[i] , 2 )
    elif cities[i] == 1 :
        fares[i] = round( 1.666667 * fares[i] , 2 )
    elif cities[i] == 2 :
        fares[i] = round( 0.85 * fares[i] , 2 )
    elif cities[i] == 3 :
        fares[i] = round( 0.8 * fares[i] , 2 )
    elif cities[i] == 4 :
        fares[i] = round( 1.0 * fares[i] , 2 )
    elif cities[i] == 5 :
        fares[i] = round( 1.666667 * fares[i] , 2 )
#print fares
df = pd.DataFrame({
    'cities' : cities ,
    'fares' : fares ,
    'distances' : distances ,
    'latitudes1' : latitudes1 ,
    'longitudes1' : longitudes1 ,
    'latitudes2' : latitudes2 ,
    'longitudes2' : longitudes2 ,
    'mode_sns' : mode_sns })
df['fares/km'] = df['fares'] / df['distances']
df.plot.scatter( x = 'mode_sns' , y = 'fares/km' , s = 1 )
to_save = pd.DataFrame({
    'TIMESTAMP' : timestamps ,
    'STARTING_LATITUDE' : latitudes1 ,
    'STARTING_LONGITUDE' : longitudes1 ,
    'DESTINATION_LATITUDE' : latitudes2 ,
    'DESTINATION_LONGITUDE' : longitudes2 ,
    'VEHICLE_TYPE'  : mode_names ,
    'TOTAL_LUGGAGE_WEIGHTS' : luggage_weights ,
    'WAIT_TIME' : wait_times ,
    'TRAFFIC_STUCK_TIME' : traffic_wait_times ,
    'DISTANCE' : distances
    
} , columns = ( 
               'TIMESTAMP' ,
               'STARTING_LATITUDE' ,
               'STARTING_LONGITUDE' ,
               'DESTINATION_LATITUDE' ,
               'DESTINATION_LONGITUDE' ,
               'VEHICLE_TYPE' ,
               'TOTAL_LUGGAGE_WEIGHTS' ,
               'WAIT_TIME' ,
               'TRAFFIC_STUCK_TIME' ,
               'DISTANCE'
              ) ,
    index = range(1,samples+1)  )
to_save.to_csv("intracity_fare_test.csv")
to_save = pd.DataFrame({
    'FARE' : fares
    
} ,
    index = range(1,samples+1)  )
to_save.to_csv("intracity_fare_sample_submission.csv")
        
        
        
        
```

```python
#Error generator
gps_err = True                     # For introducing gps error
odometer_removal_err = True        # For introducing odometer removal error
odometer_non_removal_err = True    # For odometer non removal error
unfair_fare = True                 # For unfair fare
fare_randomness = True             # Fare randomness
ibucket = [ i for i in range(0,samples) ]
itouched = []
# GPS Unavailable error - 0.50% of total data
print "GPS Error"
for _ in range(0, int( 0.05*samples ) ):
    if not gps_err:
        continue
    sel = np.random.randint( 0 , len(ibucket) )
    index = ibucket[sel]
    del ibucket[sel]
    # err_type    description
    # 0           starting gps location is missing
    # 1           destination gps location is missing
    # 2           both gps location is missing
    err_type = np.random.randint(0,3)
    if err_type == 0:
        latitudes1[index] = np.nan
        longitudes1[index] = np.nan
    elif err_type == 1:
        latitudes2[index] = np.nan
        longitudes2[index] = np.nan
    else:
        latitudes1[index] = np.nan
        longitudes1[index] = np.nan
        latitudes2[index] = np.nan
        longitudes2[index] = np.nan
    itouched.append( index )
    print err_type , index
# Odometer reading error - 2.5% of total data
# Removal error
print "Odometer removal error"
for _ in range( 0 , int( 0.0125*samples ) ):
    print index
    if not odometer_removal_err:
        continue
    sel = np.random.randint( 0 , len(ibucket) )
    index = ibucket[sel]
    del ibucket[sel]
    dist_calc = int( distance( latitudes1[index] , longitudes1[index] , latitudes2[index] , longitudes2[index] ) )
    if dist_calc == 0:
        distances[index] = 0
    else:
        distances[index] = np.random.randint( 0 , dist_calc )
    itouched.append( index )
# Non Removal error
print "Odometer non removal error"
for _ in range( 0 , int( 0.0125*samples ) ):
    if not odometer_non_removal_err:
        continue
    sel = np.random.randint( 0 , len(ibucket) )
    index = ibucket[sel]
    del ibucket[sel]
    if latitudes1[index] == np.nan or latitudes2[index] == np.nan:
        continue
    dist_calc = int( distance( latitudes1[index] , longitudes1[index] , latitudes2[index] , longitudes2[index] ) )
    distances[index] = np.random.randint( dist_calc , 100 )
    itouched.append( index )
    print index
# Unfair fare - 2.5% of total data
print "Unfair fare error"
for _ in range( 0 , int( 0.0125*samples ) ):
    if not unfair_fare:
        continue
    sel = np.random.randint( 0 , len(ibucket) )
    index = ibucket[sel]
    del ibucket[sel]
    fares[index] = np.random.randint(30,40)*distances[index] + 100
    itouched.append( index )
    print index
#Fare Randomness
for i in range(samples):
    if not fare_randomness:
        continue
    if i not in itouched :
        fares[i] =  round( ( 1.025 - np.random.random()*0.05  ) * fares[i] , 2 )
```
